```python
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import pandas as pd

start_date = datetime(2020, 1, 1)
end_date = datetime(2024, 8, 31)
stock_symbols = ['NVDA', 'AMD', 'INTC', 'QCOM', 'AAPL', 'AVGO']

all_stocks_data = pd.DataFrame()

for symbol in stock_symbols:
    stock_data = yf.download(symbol, start=start_date, end=end_date)['Adj Close']
    all_stocks_data[symbol] = stock_data

all_stocks_data.dropna(inplace=True)

plt.figure(figsize=(12, 8))
for symbol in stock_symbols:
    sns.kdeplot(all_stocks_data[symbol], linewidth=2, fill=True, label=symbol)
plt.title(f"Kernel Density Estimation of Stock Prices")
plt.xlabel("Price")
plt.ylabel("Density")
plt.legend(title="Stock Symbols")
plt.grid(True)
plt.tight_layout()

plt.show()

summary_statistics = all_stocks_data.describe()

print("Summary Statistics of the Training Period:")
print(summary_statistics)
```
